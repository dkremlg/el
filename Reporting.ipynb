```python
import numpy as np
import pandas as pd
from pandas import Series
from pandas import DataFrame

from datetime import date
```

```python
Data_2019=pd.read_csv('reporting_LUX-VIE_2019_last.csv',header=None)
Data_2019.columns=['ExtractDate','DepDate','FltNum','ArrivalAirport','DepartAirport',
                  'BookingClass','Pax','Dprio_1','Dprio_2']
Data_2019=Data_2019[['DepDate','FltNum','ArrivalAirport','DepartAirport','BookingClass','Pax']]
Data_2018=pd.read_csv('reporting_LUX-VIE_2018_last.csv',header=None)
Data_2018.columns=['ExtractDate','DepDate','FltNum','ArrivalAirport','DepartAirport',
                  'BookingClass','Pax','Dprio_1','Dprio_2']
Data_2018=Data_2018[['DepDate','FltNum','ArrivalAirport','DepartAirport','BookingClass','Pax']]
```

```python
Data_2019=pd.pivot_table(Data_2019,index=['DepDate','FltNum','ArrivalAirport','DepartAirport'],columns='BookingClass',values='Pax').reset_index()
Data_2019=Data_2019.fillna(0)
```

```python
Data_2018=pd.pivot_table(Data_2018,index=['DepDate','FltNum','ArrivalAirport','DepartAirport'],columns='BookingClass',values='Pax').reset_index()
Data_2018=Data_2018.fillna(0)
```

```python
Data_2019['DepDate']=pd.to_datetime(Data_2019['DepDate'].apply(lambda x: str(x)[0:4]+'-'+str(x)[4:6]+'-'+str(x)[6:8]))
Data_2018['DepDate']=pd.to_datetime(Data_2018['DepDate'].apply(lambda x: str(x)[0:4]+'-'+str(x)[4:6]+'-'+str(x)[6:8]))
```

```python
Data_2019['#']=Data_2019[[x for x in Data_2019.columns if len(x)==1]].apply(sum,axis=1)
Data_2018['#']=Data_2018[[x for x in Data_2018.columns if len(x)==1]].apply(sum,axis=1)
```

```python
# Data_2019['Week']=Data_2019['DepDate'].apply(lambda x: x.isocalendar()[1])
# Data_2018['Week']=Data_2018['DepDate'].apply(lambda x: x.isocalendar()[1])
Data_2019['Week']=Data_2019['DepDate'].apply(lambda x: x.month)
Data_2018['Week']=Data_2018['DepDate'].apply(lambda x: x.month)
```

```python
Data_2019=Data_2019.groupby(['FltNum','Week'])[[x for x in Data_2019.columns if len(x)==1]].sum().reset_index()
Data_2018=Data_2018.groupby(['FltNum','Week'])[[x for x in Data_2018.columns if len(x)==1]].sum().reset_index()
```

```python
allclasses=['T','I','W','R','A','N','G','O','H','V','F','X','L','U','P','S','B','Q','K','M','Y','D','Z','C','J']
allclasses=[x for x in reversed(allclasses)]
```

```python
Data_2019=Data_2019[[x for x in Data_2019.columns if len(x)>1]+[x for x in allclasses if x in Data_2019.columns]+['#']]
Data_2018=Data_2018[[x for x in Data_2018.columns if len(x)>1]+[x for x in allclasses if x in Data_2018.columns]+['#']]
```

```python
for i in [x for x in Data_2019.columns if len(x)==1 and x!='#']:
    Data_2019[i]=np.round(Data_2019[i]/Data_2019['#'],2)
```

```python
for i in [x for x in Data_2018.columns if len(x)==1 and x!='#']:
    Data_2018[i]=np.round(Data_2018[i]/Data_2018['#'],2)
```
