# Check method for RASK percentile computation

```python
Data['RASK']=Data['RASK'].astype('str').apply(lambda x: x.replace(',','.')).astype('float')
Data['Percentile_RASK_target']=Data['Percentile_RASK_target'].astype('str').apply(lambda x: x.replace(',','.')).astype('float')
Data['Load factor']=Data['Load factor'].astype('str').apply(lambda x: x.replace(',','.')).astype('float')
Data['Yield']=Data['Yield'].astype('str').apply(lambda x: x.replace(',','.')).astype('float')
Data['Sum_Revenues']=Data['Sum_Revenues'].astype('str').apply(lambda x: x.replace(',','.')).astype('float')
Data['Sum_RPK']=Data['Sum_RPK'].astype('str').apply(lambda x: x.replace(',','.')).astype('float')
percRASK_byKey=Data.groupby('key')['RASK'].apply(lambda x: np.quantile(x,0.75)).reset_index()
#Data[['key','Percentile_RASK_target']].merge(percRASK_byKey,on='key').loc[0:15,:]
```

```python
Data=Data.sort_values(by='key')

keys=Data['key'].unique().tolist()

container_rask_percentile=[]
container_NewRevenue=[]
container_RevUpside=[]
container_LF_percentile=[]
container_NewRPK=[]
container_RPKUpside=[]
container_PAXUpside=[]

for k in keys:

    intermediate=Data.loc[Data['key']==k,:].copy()

    NONE=intermediate.loc[intermediate['type']=='NONE',:].copy()
    SPOIL=intermediate.loc[intermediate['type']=='SPOIL',:].copy()
    SpoilYes=intermediate.loc[(intermediate['type']=='SPOIL')&(intermediate['uplift_population']=='YES'),:].copy()
    
    rask_percentile=np.quantile(intermediate['RASK'],0.75)
    container_rask_percentile=container_rask_percentile+[rask_percentile]*SpoilYes.shape[0]

    if SpoilYes.shape[0]>0:
    
        NewRevenue=(SpoilYes['Percentile_RASK_target'].unique()*SpoilYes['Sum_ASK'])/100
        container_NewRevenue=container_NewRevenue+NewRevenue.tolist()

        RevUpside=NewRevenue-SpoilYes['Sum_Revenues']
        container_RevUpside=container_RevUpside+RevUpside.tolist()

        if NONE.shape[0]>0:
        
            LF_percentile=np.quantile(NONE['Load factor'],0.75)
        else: 
            LF_percentile=np.quantile(SPOIL['Load factor'],0.75)
        container_LF_percentile=container_LF_percentile+[LF_percentile]*SpoilYes.shape[0]

        NewRPK=(NewRevenue*LF_percentile*100)/rask_percentile
        container_NewRPK=container_NewRPK+NewRPK.tolist()

        RPKUpside=NewRPK-SpoilYes['Sum_RPK']
        container_RPKUpside=container_RPKUpside+RPKUpside.tolist()

        PAXUpside=RPKUpside/SpoilYes['Stage Length']
        container_PAXUpside=container_PAXUpside+PAXUpside.tolist()

SpoilYes=Data.loc[(Data['type']=='SPOIL')&(Data['uplift_population']=='YES'),:]
SpoilYes.index=range(SpoilYes.shape[0])

pd.concat([SpoilYes['Percentile_RASK_target'],
          Series(container_rask_percentile),
          SpoilYes['New Revenue'],
          Series(container_NewRevenue),
          SpoilYes['Rev Upside'],
          Series(container_RevUpside),
          SpoilYes['New RPK'],
          Series(container_NewRPK),
          SpoilYes['RPK Upside'],
          Series(container_RPKUpside),
          SpoilYes['PAX Upside'],
          Series(container_PAXUpside)],axis=1)
```
